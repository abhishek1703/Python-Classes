1. Write a program that takes a list of integers and separates them into three lists: positive, negative, and zero. Then print all three lists.

```python
nums = [10, -5, 0, 7, -2, 0]
positive = []
negative = []
zero = []
for n in nums:
    if n > 0:
        positive.append(n)
    elif n < 0:
        negative.append(n)
    else:
        zero.append(n)
print(f"Positive numbers: {positive}")
print(f"Negative numbers: {negative}")
print(f"Zero numbers: {zero}")
```

2. Given a dictionary of item prices, ask the user for a budget and display which items can be bought within that budget.

```python
items = {
    "Pen": 10,
    "Book": 50,
    "Bottle": 100,
    "Bag": 500
}
budget = int(input("Enter your budget: "))
for item in items:
    if items[item] <= budget:
        print(f"You can buy {item} for {items[item]}")
```

3. Create a program that reads a string and counts vowels and consonants separately. Ignore spaces and punctuation.

```python
text = input("Enter a string: ")
vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0
for ch in text:
    if ch.isalpha():
        if ch in vowels:
            vowel_count += 1
        else:
            consonant_count += 1
print(f"Vowels = {vowel_count}")
print(f"Consonants = {consonant_count}")
```

4. Given a list of numbers, use a loop to build a new list containing only numbers whose digits sum to an even value.

```python
nums = [12, 35, 44, 81, 90]
new_list = []
for n in nums:
    total = 0
    for digit in str(n):
        total += int(digit)
    if total % 2 == 0:
        new_list.append(n)
print(f"Numbers are: {new_list}")
```

5. Write a program that takes two lists and prints the common elements, elements unique to the first list, and elements unique to the second list.

```python
list1 = [1, 2, 3, 4]
list2 = [3, 9, 5, 7]
common = []
only1 = []
only2 = []
for x in list1:
    if x in list2:
        common.append(x)
    else:
        only1.append(x)
for x in list2:
    if x not in list1:
        only2.append(x)
print(f"Common elements: {common}")
print(f"Only in first list: {only1}")
print(f"Only in second list: {only2}")
```

6. Create a nested if/elif structure to evaluate a temperature reading: print “Cold” if below 10, “Cool” if 10–20, “Warm” if 21–30, and “Hot” if above 30. If the value is outside a realistic range (below -50 or above 60), print “Sensor error”.

```python
temp = int(input("Enter temperature: "))
if temp < -50 or temp > 60:
    print("Sensor error")
elif temp < 10:
    print("Cold")
elif temp <= 20:
    print("Cool")
elif temp <= 30:
    print("Warm")
else:
    print("Hot")
```

7. Using a while loop, generate Fibonacci numbers until the next number would exceed 200. Store them in a list and print the list.

```python
fib = []
a = 0
b = 1
while a <= 200:
    fib.append(a)
    c = a + b
    a = b
    b = c
print(f"Fibonacci series: {fib}")
```

8. Given a list of words, use list comprehension to create a list of words that are palindromes and have length greater than 3.

```python
words = ["madam", "cat", "level", "python", "radar", "table"]
result = [w for w in words if w == w[::-1] and len(w) > 3]
print(f"Palindrome words: {result}")
```

```python
nums = {2, 4, 5, 7, 11, 15, 17}
prime = []
for n in nums:
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    if count == 2:
        prime.append(n)
prime.sort()
print(f"Prime numbers: {prime}")
```
