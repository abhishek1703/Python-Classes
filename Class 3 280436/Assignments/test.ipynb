```python
from collections import ChainMap

default_prices = {
    "course_fee": 10000,
    "discount": 0
}

festival_offer = {
    "discount": 2000
}

student_specific = {
    "course_fee": 8000
}

Stu_spec_prices = ChainMap(student_specific, festival_offer, default_prices)

print(Stu_spec_prices["course_fee"]) 
print(Stu_spec_prices["discount"])

print(dict(Stu_spec_prices))


```

```python
from collections import UserList

class CustomList(UserList):
    def __init__(self, items):
        if any(x < 0 for x in items):
            raise ValueError("Negative values are not allowed")
        
list = CustomList()
list.append(4);
list.append(-5)  
print(list)
```
