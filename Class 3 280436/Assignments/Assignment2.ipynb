```python
numbers = [45, 23, 89, 12, 56]
maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num

    if num < minimum:
        minimum = num

print(f'The Maximum from given numbers is {maximum}')
print(f'The Miniimum from given numbers is {minimum}')
```

6. Write a program that takes a test score (0-100) as input. Use nested if-elif-else statements to assign grades: A (90-100), B (80-89), C (below 80). Handle invalid scores (below 0 or above 100).

```python
test_score = input("Enter score (0-100): ")

score = int(test_score)

if score < 0 or score > 100:
    print("Invalid Score")
elif score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Grade: C")
```

7. Use a while loop to iterate from 1 to 10. For each number, check if it's even. If even, square it; if odd, skip it. Calculate and print the total sum of all squared even numbers.

```python
num = 1
total = 0
while num <= 10:
    
    if num % 2 == 0:
        square = num ** 2
        total = square + total
    
    num += num

print("The sum of even numbers are", total)
```

9. You have a dictionary containing a list of students with their names and scores. For each student, calculate their average score. Based on the average, classify them as 'Excellent' (≥90), 'Good' (80-89), or 'Average' (<80). Store the results in a new dictionary and display them.

```python
students = {
    "Rahul": [95, 92, 90],
    "Aman": [85, 80, 82],
    "Neha": [70, 75, 78]
}
results = {}
for name, scores in students.items():
    average = sum(scores) / len(scores)
    if average >= 90:
        category = "Excellent"
    elif average >= 80:
        category = "Good"
    else:
        category = "Average"
    results[name] = {
        "Average": average,
        "Category": category
    }
print(results)
```
