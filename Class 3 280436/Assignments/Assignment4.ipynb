Q14. Write a function count_even_odd(*numbers) that returns how many even and
odd numbers are passed.
Example:
count_even_odd(1, 2, 3, 4, 5, 6)
Expected output:
Even: 3, Odd: 3

```python
def count_even_odd(*numbers):
    even_count = 0
    odd_count = 0

    for num in numbers:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

    return f"Even: {even_count}, Odd: {odd_count}"
print(count_even_odd(1, 2, 3, 4, 5, 6))
```

Q20. Write a function using the structure below:

def ratio(numerator, denominator, /, *, as_percent=False):
The function should return the normal ratio when as_percent=False and
percentage when as_percent=True .
Example:
ratio(1, 4)
Expected output:
0.25
Example:
ratio(1, 4, as_percent=True)
Expected output:
25.0

```python
def ratio(numerator, denominator, /, *, as_percent=False):
    result = numerator / denominator

    if as_percent:
        return result * 100

    return result

print(ratio(1, 4))

print(ratio(1, 4, as_percent=True))
```

Q23. Write a function calculate_final_price(price, tax_rate=0.05, discount=0) that
subtracts discount first and then applies tax.

```python
def calculate_final_price(price, tax_rate=0.05, discount=0):
    price_after_discount = price - discount
    final_price = price_after_discount + (price_after_discount * tax_rate)
    return final_price

print(calculate_final_price(1000))
```

Q29. Write a function with the following structure:

def average(*numbers: float) -> float:
The function should return the average of all numbers. If no number is passed,
return 0.0 .

```python
def average(*numbers: float) -> float:
    if len(numbers) == 0:
        return 0.0

    return sum(numbers) / len(numbers)

print(average(10, 20, 30))
print(average(5, 10))
print(average())
```
